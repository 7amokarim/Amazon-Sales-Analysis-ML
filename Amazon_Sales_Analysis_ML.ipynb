```python
import pandas as pd
import numpy as np

# Convert relevant columns to numeric types
# discounted_price
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

# actual_price
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

# discount_percentage
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# rating
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# rating_count
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

5️⃣ Data Cleaning and Feature Engineering
a. Handling Missing Values

```python
# Check missing values
df.isnull().sum()

# Identify numerical columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Fill numerical missing values with mean of numerical columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Drop rows with missing categorical values
# Assuming 'Region' might not exist or be fully populated after initial data load, focusing on 'category'
# If other categorical columns need specific handling, they should be listed here.
df.dropna(subset=['category'], inplace=True)
```

b. Encoding categorical variables

```python
df_encoded = pd.get_dummies(df, columns=['category'], drop_first=True)
df_encoded.head()
```

c. Feature Engineering

```python
# Example: Total Sale = Price * Quantity
# The columns 'Amount' and 'Qty' were not found in the DataFrame.
# Assuming 'Amount' refers to 'discounted_price'.
# A 'Quantity' column is essential for calculating 'Total_Sale' as 'Price * Quantity'.
# Since no 'Qty' column was identified, we will use a placeholder of 1 for quantity.
# Please ensure you have a 'Qty' column or specify how quantity should be derived if this is not the intended calculation.
df_encoded['Total_Sale'] = df_encoded['discounted_price'] * 1
df_encoded.head()
```

8️ Hypothesis Testing (Significance Test)

```python
from scipy.stats import ttest_ind

# Split data: below median price vs above median price
median_price = df_encoded['discounted_price'].median()
low_price_sales = df_encoded[df_encoded['discounted_price'] <= median_price]['Total_Sale']
high_price_sales = df_encoded[df_encoded['discounted_price'] > median_price]['Total_Sale']

# t-test
t_stat, p_val = ttest_ind(low_price_sales, high_price_sales)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

##  Statistical Tests (T-Test & ANOVA)
In this section, we apply statistical hypothesis testing to check if there are significant differences between groups.  
We test:
- Independent t-test between high-sale and low-sale groups.
- ANOVA test to compare multiple product categories.
These tests help us understand whether differences in sales patterns are statistically meaningful.


```python
# ========== Statistical Test 1: Independent T-Test ==========

from scipy.stats import ttest_ind

# Split dataset into high-sale and low-sale groups
median_sale = df['Total_Sale'].median()

high_sale = df[df['Total_Sale'] > median_sale]['Total_Sale']
low_sale = df[df['Total_Sale'] <= median_sale]['Total_Sale']

# Perform t-test
t_stat, p_value = ttest_ind(high_sale, low_sale, equal_var=False)

print("T-Test: High Sale vs Low Sale")
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

ANOVA

```python

# ========== Statistical Test 2: ANOVA for Categories ==========

from scipy.stats import f_oneway

if cat_col:
    grouped = [group["Total_Sale"].values for name, group in df.groupby(cat_col)]
    f_stat, p_val = f_oneway(*grouped)

    print("ANOVA Test Across Categories")
    print("F-Statistic:", f_stat)
    print("P-Value:", p_val)
else:
    print("Category column not available for ANOVA.")
```

##  Feature Encoding & Train/Test Split
We transform categorical columns into numeric form, scale the numerical features, and split the dataset into training and testing sets.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Copy dataset
df_ml = df.copy()

# Identify categorical columns for Label Encoding
# 'category' is the primary categorical column that was one-hot encoded in df_encoded.
# For df_ml, we'll use LabelEncoder on it.
# Other object columns like product_id, product_name, review_content are typically dropped or require different text processing.
cat_cols = ['category']

# Encode categorical columns
for col in cat_cols:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col].astype(str))

# Select features (everything except target)
# Corrected 'total_sale' to 'Total_Sale'
X = df_ml.drop("Total_Sale", axis=1)

# Corrected 'total_sale' to 'Total_Sale'
y = df_ml["Total_Sale"]

# Dropping columns that are unsuitable for direct ML model input or high cardinality
# These include unique identifiers and free-text fields.
X = X.drop(columns=['product_id', 'product_name', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link'], errors='ignore')

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

X_train.shape, X_test.shape
```

##  Model 1: Linear Regression
A baseline model used to understand linear relationships between features and total sales.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np # Import numpy for sqrt function

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

print("Linear Regression Results")
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print("R2 Score:", r2_score(y_test, y_pred_lr))
```

## Model 2: Random Forest Regressor
A more powerful non-linear model capable of capturing complex patterns.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np # Ensure numpy is imported for sqrt

rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Random Forest Results")
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("R2 Score:", r2_score(y_test, y_pred_rf))
```

Feature Importance
We examine which variables have the strongest impact on predicted sales.


```python
import numpy as np

importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = X.columns

plt.figure(figsize=(10,6))
sns.barplot(x=importances[indices], y=feature_names[indices])
plt.title("Feature Importance (Random Forest)")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "feature_importance_rf.png"))
plt.show()
```

##  Model Comparison
We compare RMSE and R2 score between all trained models.


```python
import pandas as pd
import numpy as np # Ensure numpy is imported for sqrt
from sklearn.metrics import mean_squared_error, r2_score

results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "RMSE": [
        np.sqrt(mean_squared_error(y_test, y_pred_lr)),
        np.sqrt(mean_squared_error(y_test, y_pred_rf))
    ],
    "R2 Score": [
        r2_score(y_test, y_pred_lr),
        r2_score(y_test, y_pred_rf)
    ]
})

results
```
